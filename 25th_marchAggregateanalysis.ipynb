```python
# data aggregation using single criteria for summarizing values 
# what is the total basic salary for delta state employees
deltasumif = salah.query('STATE == "DELTA"')["BASIC SALARY"].sum()
deltasumif
```

```python
#Aggregating multiple columns using groupby function
# summarizing categorical data using groupby()
grouptot = salah.groupby(by = ["DEPT"])["BASIC_SALARY"].sum().reset_index()
grouptot
```

```python
sns.barplot(x = "BASIC_SALARY",y ="DEPT",data =grouptot,palette = "bright")
plt.title("BASIC SALARY BY DEPT",fontsize =16)
plt.xlabel("BASIC_SALARY",fontsize = 12)
plt.ylabel("DEPT",fontsize =12)
plt.xticks(rotation=45)
plt.show
```
